```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# model names
models = ['CNN', 'CRNN', 'LSTM', 'LSTM+CNN']

# accuracy values
word2vec = [61.9, 63.24, 62, 68.9]
openai = [67.45, 70.45, 69.34, 73.45]

x = np.arange(len(models))  # label locations
width = 0.35  # width of the bars

fig, ax = plt.subplots()

# plot bars
rects1 = ax.bar(x - width/2, word2vec, width, label='Word2Vec')
rects2 = ax.bar(x + width/2, openai, width, label='OpenAI')

# add some text for labels, title
ax.set_ylabel('Accuracy')
ax.set_title('Accuracy by model and embedding')
ax.set_xticks(x)
ax.set_xticklabels(models)
ax.legend(bbox_to_anchor=(1.05, 1), loc='upper left')

# function to attach a text label above each bar displaying its height
def autolabel(rects):
    for rect in rects:
        height = rect.get_height()
        ax.annotate('{}'.format(height),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')

autolabel(rects1)
autolabel(rects2)

fig.tight_layout()
plt.show()
```

```python
# accuracy values
word2vec = [61.9, 63.24, 62, 68.9]
openai = [67.45, 70.45, 69.34, 73.45]

# calculate differences
differences = [o - w for o, w in zip(openai, word2vec)]

# calculate average improvement
average_improvement = sum(differences) / len(differences)

print("Average improvement with OpenAI embedding: ", average_improvement)
```
