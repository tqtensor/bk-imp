```python
import random

import matplotlib.pyplot as plt
import networkx as nx
import walker
```

```python
# Create three graphs with 10 nodes each
G1 = nx.erdos_renyi_graph(10, p=0.3, seed=42)
G2 = nx.erdos_renyi_graph(10, p=0.3, seed=43)

# Combine G1 and G2 into a single graph G
G = nx.disjoint_union(G1, G2)

# Find two nodes that are not connected
non_connected_nodes = list(nx.non_edges(G))
node1, node2 = random.choice(non_connected_nodes)

# Add an edge to connect the two nodes
G.add_edge(node1, node2)

# Remove node that connected to just one edge
nodes_to_remove = [
    node for node, degree in dict(G.degree()).items() if degree == 1
]
G.remove_nodes_from(nodes_to_remove)

# Reset node labels, because some nodes were removed
G = nx.convert_node_labels_to_integers(G)

# Initialize node labels with unique values
initial_labels = {node: node for node in G.nodes}

# Visualize the graph
pos = nx.spring_layout(G, k=0.3)
plt.figure(figsize=(8, 8))
nx.draw(
    G,
    pos,
    with_labels=True,
    node_color=list(initial_labels.values()),
    cmap=plt.cm.RdYlBu,
    font_color="white",
    node_size=500,
    edge_color="gray",
    width=1.5,
)
plt.title("Combined Graph")
plt.show()
```

```python
# Label propagation function
paths = walker.random_walks(G, n_walks=10, walk_len=len(G.nodes), alpha=.1)
paths = [list(path) for path in paths]

def label_propagation(graph, initial_labels, random_walks, max_iterations):
    labels = initial_labels.copy()
    for iteration in range(min(len(random_walks), max_iterations)):
        explanation = []
        path = random_walks[iteration]
        for node in path:
            explanation.append(f"Node {node} has label {labels[node]}")
            neighbors = list(graph.neighbors(node))
            explanation.append(
                f", has neighbors {neighbors} with labels {[labels[neighbor] for neighbor in neighbors]}"
            )
            if neighbors:
                # Update the label based on the majority vote of neighbors
                neighbor_labels = [labels[neighbor] for neighbor in neighbors]
                majority_label = max(
                    set(neighbor_labels), key=neighbor_labels.count
                )
                explanation.append(
                    f", majority label is {majority_label}, so set label of node {node} to {majority_label}\n"
                )
                labels[node] = majority_label
        print("Iteration {}: {}".format(iteration + 1, "".join(explanation)))

        # Visualize the graph at each iteration
        plt.figure(figsize=(8, 8))
        nx.draw(
            graph,
            pos,
            with_labels=True,
            node_color=list(labels.values()),
            cmap=plt.cm.RdYlBu,
            font_color="white",
            node_size=500,
            edge_color="gray",
            width=1.5,
        )
        plt.title(f"Iteration {iteration + 1}")
        plt.show()


# Run label propagation and visualize each iteration
label_propagation(G, initial_labels, paths, 15)
```
